```python
from dbfread import DBF
import pandas as pd
from pandas import DataFrame
```

```python
dbf5= DBF('05players.DBF')
dbf6= DBF('06players.DBF')
dbf7= DBF('07players.DBF')
dbf8= DBF('08players.DBF')
```

```python
df5 = DataFrame(iter(dbf5))
df6 = DataFrame(iter(dbf6))
df7 = DataFrame(iter(dbf7))
df8 = DataFrame(iter(dbf8))
```

```python
df_p5 = df5.loc[:,('NAME','FNAME','POSITION','POSITION2','HAND','OVERALLRTG',
          'FGPBASE','THREEPTBAS','FTPBASE','DNKABILITY','STLABILITY',
          'BLKABILITY','OREABILITY','DREABILITY','BALABILITY','OFFABILITY',
          'DEFABILITY','SPEED','QUICK','JUMP','DRIBBLE','DSTRENGTH',
          'DHARDY','DSHOOTRANG','FATIGUE','INSIDESC','THREEPTATT',
          'PRIMACY','CHEMTYPE'
         )]
df_p5['year'] = '05'
df_p6 = df6.loc[:,('NAME','FNAME','POSITION','POSITION2','HAND','OVERALLRTG',
          'FGPBASE','THREEPTBAS','FTPBASE','DNKABILITY','STLABILITY',
          'BLKABILITY','OREABILITY','DREABILITY','BALABILITY','OFFABILITY',
          'DEFABILITY','SPEED','QUICK','JUMP','DRIBBLE','DSTRENGTH',
          'DHARDY','DSHOOTRANG','FATIGUE','INSIDESC','THREEPTATT',
          'PRIMACY','CHEMTYPE'
         )]
df_p6['year'] = '06'
df_p7 = df7.loc[:,('NAME','FNAME','POSITION','POSITION2','HAND','OVERALLRTG',
          'FGPBASE','THREEPTBAS','FTPBASE','DNKABILITY','STLABILITY',
          'BLKABILITY','OREABILITY','DREABILITY','BALABILITY','OFFABILITY',
          'DEFABILITY','SPEED','QUICK','JUMP','DRIBBLE','DSTRENGTH',
          'DHARDY','DSHOOTRANG','FATIGUE','INSIDESC','THREEPTATT',
          'PRIMACY','CHEMTYPE'
         )]
df_p7['year'] = '07'
df_p8 = df8.loc[:,('NAME','FNAME','POSITION','POSITION2','HAND','OVERALLRTG',
          'FGPBASE','THREEPTBAS','FTPBASE','DNKABILITY','STLABILITY',
          'BLKABILITY','OREABILITY','DREABILITY','BALABILITY','OFFABILITY',
          'DEFABILITY','SPEED','QUICK','JUMP','DRIBBLE','DSTRENGTH',
          'DHARDY','DSHOOTRANG','FATIGUE','INSIDESC','THREEPTATT',
          'PRIMACY','CHEMTYPE'
         )]
df_p8['year'] = '08'
```

```python
df_p = pd.concat([df_p5, df_p6, df_p7, df_p8], ignore_index=True)
```

```python
df_p.to_csv('player_cat.csv', sep=',')
```
